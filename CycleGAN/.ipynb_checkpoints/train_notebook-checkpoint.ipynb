```python
#set training parameters

from models import Notebook_GanConfig
import shutil
from utils import *
import os

phase = 'train'
light = True
# define dataset
dataset = 'face2cartoon'

# training parameters
iteration = 60000
batch_size = 1
print_freq = 1000
save_freq = 30000

decay_flag = True
# learning rate
lr = 0.0001
adv_weight = 1
cycle_weight = 50
identity_weight = 10
cam_weight = 1000
faceid_weight = 1
ch = 32
n_dis = 6
# input image size
img_size = 256
# input image channel
img_ch = 3
# use GPU to train
device = 'cuda:0'

benchmark_flag = False
resume = False
rho_clipper = 1.0
w_clipper = 1.0
pretrained_weights ='train_model/train-size256-ch32-True-lr0.0001-adv1-cyc50-id1-identity10-cam1000/face2cartoon_params_latest(pretrained).pt'
```

```python
result_dir = './train_model/train-size{}-ch{}-{}-lr{}-adv1-cyc50-id1-identity10-cam1000'.format(
        img_size,
        ch,
        light,
        lr)
```

```python
def check_args():
    check_folder(os.path.join(result_dir, dataset, 'model'))
    check_folder(os.path.join(result_dir, dataset, 'img'))
    check_folder(os.path.join(result_dir, dataset, 'test'))
    shutil.copy(__file__, result_dir)
    return args
```

```python
#run this cell for training! load pretrained model and train for another 60000 epoch
gan = Notebook_GanConfig(light, result_dir, dataset, iteration, decay_flag, batch_size, print_freq, save_freq, lr, ch, 
                 adv_weight, cycle_weight, identity_weight, cam_weight, faceid_weight, n_dis, img_size, img_ch, 
                 device, benchmark_flag, resume, rho_clipper, w_clipper, pretrained_weights)

gan.build_model()

if phase == 'train':
    gan.train()
    print(" [*] Training finished!")

if phase == 'test':
    gan.test()
    print(" [*] Test finished!")
```

```python
#run this cell for test!

import os
import cv2
from models.Test_Config import Face2Cartoon
#from skimage import io

if __name__ == '__main__':
    Input_path = './test_images/'
    Save_path = './save_images/'
    filename_list = os.listdir(Input_path)
    for i in filename_list:
        save_name = i.split('.')[0] + '.png'       
        img = cv2.cvtColor(cv2.imread(f'./test_images/{i}'), cv2.COLOR_BGR2RGB)
        c2p = Face2Cartoon()
        cartoon = c2p.inference(img)
        if cartoon is not None:
            cv2.imwrite(Save_path + save_name, cartoon)
```
